```python
df=pd.read_csv('new_algerian.csv')
df.drop(columns=['Unnamed: 0'], inplace=True)
df.head()
```

```python
df.columns
```

```python
df['Classes']=np.where(df['Classes'].str.contains('not fire'), 'not fire', 'fire')
```

```python
# Monthly Fire Analysis

dftemp = df.loc[df['Region']==1]
plt.subplots(figsize=(12,6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=df)
plt.ylabel('Number of fires', weight='bold')
plt.xlabel('Months', weight='bold')
plt.title("Fire Analysis of Sidi-Bel Regions", weight='bold')
```

```python
# Monthly Fire Analysis 
dtemp = df.loc[df['Region']==0]
plt.subplots(figsize=(13,6))
sns.set_style('darkgrid')
sns.countplot(x='month', hue='Classes', data=df)
plt.xlabel('Months', weight='bold')
plt.ylabel('Number of fires', weight='bold')
plt.title('Fire Analysis of Brjaia Regions', weight='bold')
```

##### It is observed that August and September had the most number of foreset fires for both regions. And from the abve plot of months, we can understand few things.
##### Most of the fires happened in August and very few high fires happened in only 3 months - June and July and August.
